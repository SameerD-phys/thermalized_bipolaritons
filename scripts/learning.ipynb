```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
Ebar = 1.50   # (EX + EC)/2 in eV
g = 0.06      # exciton-cavity coupling per exciton (eV)
J = 0.03      # exciton-exciton coupling between any pair (eV)
N = 3

# Detuning δ = EX - EC
dE = np.linspace(-0.5, 0.5, 800)

# Reconstruct EX and EC from Ebar and δ
EX = Ebar + dE/2
EC = Ebar - dE/2

# Exciton eigen-energies for all-to-all J (complete graph)
EB = EX + (N - 1) * J          # bright/symmetric exciton energy = EX + 2J
ED = EX - J                    # dark exciton energy (degeneracy N-1 = 2)

# Effective collective coupling to cavity
g_eff = np.sqrt(N) * g         # = sqrt(3) g

# Polariton branches from bright-cavity hybridization
E_plus  = 0.5 * (EB + EC) + np.sqrt(((EB - EC) / 2)**2 + g_eff**2)
E_minus = 0.5 * (EB + EC) - np.sqrt(((EB - EC) / 2)**2 + g_eff**2)

# --- Plot ---
plt.figure()
plt.plot(dE, E_plus,  label=r"$E_+$ (polariton)")
plt.plot(dE, E_minus, label=r"$E_-$ (polariton)")

# Two degenerate dark states (draw twice so you remember there are 2)
for k in range(N - 1):
    plt.plot(dE, ED, linestyle="--", linewidth=1, label="Dark state" if k == 0 else None)

# Optional reference lines
plt.plot(dE, EX, linestyle=":", linewidth=1, label="Bare exciton $E_X$")
plt.plot(dE, EC, linestyle=":", linewidth=1, label="Bare cavity $E_C$")

plt.axvline(0, linestyle="--", linewidth=1)
plt.xlabel(r"Detuning $\delta = E_X - E_C$ (eV)")
plt.ylabel("Energy (eV)")
plt.title("Three Excitons (all-to-all J) Collectively Coupled to One Cavity")
plt.legend()
plt.tight_layout()
plt.show()

# --- Splitting at zero detuning (between polaritons) ---
idx0 = np.argmin(np.abs(dE))
print(f"Polariton splitting at δ≈0: Δ = E_+ - E_- = {E_plus[idx0] - E_minus[idx0]:.6f} eV")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
Ebar = 1.50      # (EX + EC)/2 in eV
g = 0         # exciton-cavity coupling (same for each exciton) in eV

# Non-uniform all-to-all exciton couplings (eV)
J12 = 0.030
J13 = 0.015
J23 = 0.020

# Detuning range: δ = EX - EC
dE = np.linspace(-0.5, 0.5, 800)

# Reconstruct EX and EC from Ebar and δ
EX = Ebar + dE/2
EC = Ebar - dE/2

# Storage for eigenvalues (4 branches)
evals_all = np.empty((dE.size, 4), dtype=float)

for i in range(dE.size):
    H = np.array([
        [EX[i], J12,  J13,  g ],
        [J12,  EX[i], J23,  g ],
        [J13,  J23,  EX[i], g ],
        [g,    g,    g,    EC[i]]
    ], dtype=float)

    # Eigenvalues of symmetric real matrix
    w = np.linalg.eigvalsh(H)
    evals_all[i, :] = np.sort(w)  # sort low -> high for smooth-ish branches

# --- Plot ---
plt.figure()
for k in range(4):
    plt.plot(dE, evals_all[:, k], label=fr"Eigen-energy {k+1}")

# Optional reference lines (bare exciton & cavity)
plt.plot(dE, EX, linestyle=":", linewidth=1, label="Bare exciton $E_X$")
plt.plot(dE, EC, linestyle=":", linewidth=1, label="Bare cavity $E_C$")

plt.axvline(0, linestyle="--", linewidth=1)
plt.xlabel(r"Detuning $\delta = E_X - E_C$ (eV)")
plt.ylabel("Energy (eV)")
plt.title("3 Excitons (non-uniform all-to-all J) Coupled to 1 Cavity")
plt.legend()
plt.tight_layout()
plt.show()

# --- Print polariton splitting at ~zero detuning (between top and bottom eigenvalues) ---
idx0 = np.argmin(np.abs(dE))
splitting = evals_all[idx0, -1] - evals_all[idx0, 0]
print(f"Total spread at δ≈0 (highest - lowest eigenenergy): {splitting:.6f} eV")

# If you specifically want the UP-LP splitting (usually the two eigenvalues
# with strongest cavity content), you'd compute eigenvectors and cavity weights.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
Ebar = 1.50
g    = 0.05
hbar = 0.6582   # eV·fs

dE = 0.0
EC = Ebar - dE/2
EX = Ebar + dE/2

# --- Hamiltonian: {cavity, exciton1, exciton2} ---
H = np.array([
    [EC,  g,   g  ],
    [g,   EX,  0.0],
    [g,   0.0, EX ],
])

eigvals, eigvecs = np.linalg.eigh(H)

# --- Time evolution ---
t    = np.linspace(0, 150, 500)
psi0 = np.array([1.0, 0.0, 0.0], dtype=complex)

coeffs = eigvecs.conj().T @ psi0                      # shape (3,)
phases = np.exp(-1j * np.outer(eigvals, t) / hbar)    # shape (3, len(t))
psi_t  = eigvecs @ (coeffs[:, None] * phases)         # shape (3, len(t))

# --- Populations ---
P_cavity = np.abs(psi_t[0])**2
P_X1     = np.abs(psi_t[1])**2
P_X2     = np.abs(psi_t[2])**2
P_bright = P_X1 + P_X2

# --- Plot ---
fig, ax = plt.subplots(figsize=(14, 12))
ax.plot(t, P_cavity, label="Cavity",          linewidth=2)
# ax.plot(t, P_bright, label="Exciton (total)", linewidth=2, linestyle='--')

omega_rabi = np.sqrt(2) * g / hbar
T_rabi     = 2 * np.pi / omega_rabi
ax.axvline(T_rabi, color='gray', linestyle=':', linewidth=1)
ax.text(T_rabi + 2, 0.95, f'$T_R$ = {T_rabi:.1f} fs', fontsize=10, color='gray')

ax.set_xlabel("Time (fs)", fontsize=13)
ax.set_ylabel("Population", fontsize=13)
ax.legend(fontsize=11)
ax.set_ylim(-0.05, 1.05)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def build_H(EX: float, EC: float, g: float, J: np.ndarray) -> np.ndarray:
    J = np.asarray(J, dtype=float)
    N = J.shape[0]
    if J.shape != (N, N):
        raise ValueError("J must be a square (N,N) matrix.")
    if not np.allclose(J, J.T, atol=1e-12):
        raise ValueError("J must be symmetric (J == J.T).")
    if not np.allclose(np.diag(J), 0.0, atol=1e-12):
        raise ValueError("J diagonal entries must be 0 (no self-coupling).")
    H = np.zeros((N + 1, N + 1), dtype=float)
    H[:N, :N] = J + np.eye(N) * EX
    H[N, N] = EC
    H[:N, N] = g
    H[N, :N] = g
    return H

def spectrum_vs_detuning(
    N: int,
    Ebar: float,
    g: float,
    J: np.ndarray,
    dE_min: float = -0.5,
    dE_max: float = 0.5,
    n_pts: int = 800
):
    J = np.asarray(J, dtype=float)
    if J.shape != (N, N):
        raise ValueError(f"J must have shape ({N},{N}).")
    dE = np.linspace(dE_min, dE_max, n_pts)
    EX = Ebar + dE / 2
    EC = Ebar - dE / 2
    evals = np.empty((n_pts, N + 1), dtype=float)
    for i in range(n_pts):
        H = build_H(EX=EX[i], EC=EC[i], g=g, J=J)
        w = np.linalg.eigvalsh(H)
        evals[i, :] = np.sort(w)
    return dE, evals, EX, EC

def generate_random_J(N, J_min=0.01, J_max=0.04, seed=None):
    if seed is not None:
        np.random.seed(seed)
    J = np.random.uniform(J_min, J_max, size=(N, N))
    J = (J + J.T) / 2.0
    np.fill_diagonal(J, 0.0)
    return J

# ---------------------- USER SETTINGS ----------------------
N = 4
Ebar = 1.50
g = 0.2
J = generate_random_J(N, J_min=0.01, J_max=0.4, seed=1)
dE_min, dE_max, n_pts = -0.5, 0.5, 900
# -----------------------------------------------------------

dE, evals, EX_line, EC_line = spectrum_vs_detuning(
    N=N, Ebar=Ebar, g=g, J=J, dE_min=dE_min, dE_max=dE_max, n_pts=n_pts
)

# ── Figure 1: Eigenspectrum ─────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(14, 12), sharex=True)

for k in range(N + 1):
    ax.plot(dE, evals[:, k], label=fr"$E_{{{k+1}}}$")
ax.plot(dE, EX_line, linestyle=":", linewidth=1, color="k", label=r"Bare $E_X$")
ax.plot(dE, EC_line, linestyle=":", linewidth=1, color="gray", label=r"Bare $E_C$")
ax.axvline(0, linestyle="--", linewidth=0.8, color="gray")
ax.set_ylabel("Energy (eV)")
ax.set_title(f"{N} Excitons Coupled to 1 Cavity — Eigenspectrum")
ax.legend(fontsize=7, ncol=3, loc="best")
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 20, 2000)
dt = t[1] - t[0]

signals = [
    {
        "label": "Exponential decay",
        "formula": r"$e^{-0.5t}$",
        "signal": np.exp(-0.5 * t),
    },
    {
        "label": "Sine",
        "formula": r"$\sin(2\pi \cdot 1.5 \cdot t)$",
        "signal": np.sin(2 * np.pi * 1.5 * t),
    },
    {
        "label": "Cosine",
        "formula": r"$\cos(2\pi \cdot 1.5 \cdot t)$",
        "signal": np.cos(2 * np.pi * 1.5 * t),
    },
    {
        "label": "Sine with exponential envelope",
        "formula": r"$e^{-0.5t}\sin(2\pi \cdot 1.5 \cdot t)$",
        "signal": np.exp(-0.5 * t) * np.sin(2 * np.pi * 1.5 * t),
    },
    {
        "label": "Gaussian",
        "formula": r"$e^{-(t-5)^2/2}$",
        "signal": np.exp(-((t - 5) ** 2) / 2),
    },
]

fig, axes = plt.subplots(5, 2, figsize=(14, 18))

for i, s in enumerate(signals):
    sig = s["signal"]

    freqs = np.fft.rfftfreq(len(sig), d=dt)
    fft_vals = np.fft.rfft(sig)
    psd = np.abs(fft_vals)   # amplitude spectrum

    # --- Time domain ---
    ax_t = axes[i, 0]
    ax_t.plot(t, sig, lw=2, color='steelblue')
    ax_t.set_title(f"{s['label']}:  {s['formula']}", fontsize=13)
    ax_t.set_xlabel(r"$t$", fontsize=13)
    ax_t.set_ylabel(r"$f(t)$", fontsize=13)
    ax_t.tick_params(labelsize=11)

    # --- Frequency domain ---
    ax_f = axes[i, 1]
    ax_f.plot(freqs, psd, lw=2, color='darkorange')
    ax_f.set_xlim(0, 4)
    ax_f.set_title(f"Fourier transform:  {s['formula']}", fontsize=13)
    ax_f.set_xlabel(r"$f$", fontsize=13)
    ax_f.set_ylabel(r"$|\hat{f}(f)|$", fontsize=13)
    ax_f.tick_params(labelsize=11)

plt.tight_layout()
plt.show()
```
